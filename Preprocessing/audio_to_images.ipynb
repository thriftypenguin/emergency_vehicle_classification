```python
# Road wav files to png

audio_folder = "Data/Road Noises"
output_folder = "Data/spectograms/road"

for filename in os.listdir(audio_folder):
    audio_path = os.path.join(audio_folder, filename)
    audio_data, sample_rate = librosa.load(audio_path)
    mel_spec = librosa.feature.melspectrogram(y=audio_data, sr=sample_rate, n_mels=128, hop_length=512, n_fft=2048)
    log_mel_spec = librosa.amplitude_to_db(mel_spec, ref=np.max)
    output_path = os.path.join(output_folder, filename.replace(".wav", ".png"))
    plt.imsave(output_path, log_mel_spec, cmap='GnBu_r')
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

spectrogram_dir = "Data/spectograms/ambulance"

# List all PNG files in the directory
png_files = [f for f in os.listdir(spectrogram_dir) if f.endswith(".png")]

# Visualize a few PNG spectrograms
num_images_to_visualize = 3

for i in range(num_images_to_visualize):
    img_path = os.path.join(spectrogram_dir, png_files[i])
    img = mpimg.imread(img_path)

    plt.figure(figsize=(10, 4))
    plt.imshow(img, aspect="auto", cmap='gray')
    plt.title(f"Spectrogram {i+1}")
    plt.axis("off")
    plt.show()
```
