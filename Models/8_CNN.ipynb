## Imports

```python
import pandas as pd
import numpy as np
import random
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from keras import models
from keras import layers
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
import scipy as sp
import warnings

warnings.simplefilter("ignore")

from sklearn.model_selection import RandomizedSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
import shap
```

## Define Data Directories and Loading Function

```python
# Load CSVs into pandas dataframes (we will not train on these but use it to split the images into right group)
df_test = pd.read_csv('../data/processed_data/test.csv')
df_train = pd.read_csv('../data/processed_data/train.csv')
df_val = pd.read_csv('../data/processed_data/validation.csv')

# Define directories for images
dir_pos = '../data/spectograms/ambulance/'
dir_neg = '../data/spectograms/road/'
```

```python
# Function to load images from given dataframe and directory
def load_images(df, dir_pos, dir_neg, padding = False):
    images = []
    labels = []
    for index, row in df.iterrows():
        # Get filename and replace '.wav' with '.png'
        filename = row['filename'].replace('.wav', '.png')
        img_path_pos = dir_pos + filename
        img_path_neg = dir_neg + filename
        # Check if the image exists in the positive directory
        if os.path.exists(img_path_pos):
            if padding:            
                img = tf.image.resize_with_pad(load_img(img_path_pos), target_height = 128, target_width = 800)
            else:
                img = load_img(img_path_pos, target_size=(128, 200))
            img = img_to_array(img)
            images.append(img)
            labels.append(1.0)  # Label for positive case
        # Check if the image exists in the negative directory
        elif os.path.exists(img_path_neg):
            if padding:            
                img = tf.image.resize_with_pad(load_img(img_path_neg), target_height = 128, target_width = 800)
            else:
                img = load_img(img_path_neg, target_size=(128, 200))
            img = img_to_array(img)
            images.append(img)
            labels.append(0.0)  # Label for negative case
    # Return images and labels as numpy arrays
    return np.array(images), np.array(labels)
```

## Define Function to Build Model

```python
def build_model(pad = False,
                filters=[64, 48],
                kernel_size=(5, 5),
                strides=(2, 2),
                activation='relu',
                pool_size=(2, 2),
                dropout_rate= 0.5,
                hidden_layer_sizes=[950],
                learning_rate=0.001):

    model = tf.keras.Sequential()

    for i in range(len(filters)):
        model.add(tf.keras.layers.Conv2D(filters=filters[i],
                                         kernel_size=kernel_size,
                                         strides=strides,
                                         padding='same',
                                         data_format='channels_last',
                                         name=f'conv_{i+1}',
                                         activation=activation))

        model.add(tf.keras.layers.MaxPool2D(pool_size=pool_size, name=f'pool_{i+1}'))
        model.add(tf.keras.layers.Dropout(rate=dropout_rate))

    model.add(tf.keras.layers.Flatten())

    for i, units in enumerate(hidden_layer_sizes):
        model.add(tf.keras.layers.Dense(units=units, name=f'fc_{i+1}', activation=activation))
        model.add(tf.keras.layers.Dropout(rate=dropout_rate))

    # Final dense layer with sigmoid activation
    model.add(tf.keras.layers.Dense(units= 1, name=f'fc_{len(hidden_layer_sizes)+1}', activation='sigmoid'))

    opt = tf.keras.optimizers.legacy.Adam(learning_rate=learning_rate)

    model.compile(optimizer=opt,
                  loss=tf.keras.losses.BinaryCrossentropy(from_logits=False),
                  metrics=['accuracy', tf.keras.metrics.Precision()])

    model.build(input_shape = (None, 128, 800, 3) if pad else (None, 128, 200, 3))

    return model
```

## Load Data

```python
# Load images and labels for each split
X_train, Y_train = load_images(df_train, dir_pos, dir_neg)
X_test, Y_test = load_images(df_test, dir_pos, dir_neg)
X_val, Y_val = load_images(df_val, dir_pos, dir_neg)

#normalize pixel values
X_train = X_train/255.0
X_test = X_test/255.0
X_val = X_val/255.0
```

```python
#review first image

plt.imshow(array_to_img(X_train[0]))
plt.show()
```

## Hyper Param Search

```python
# Define the search parameters
param_distributions = {
    'filters': [[64, 48], [32, 32]],
    'kernel_size': [(5, 5), (3, 3)],
    'strides': [(1, 1), (2, 2)],
    'pool_size': [(2, 2), (3, 3)],
    'activation': ['relu', 'tanh'],
    'dropout_rate': sp.stats.uniform(scale = 1),
    'hidden_layer_sizes': [[500], [1000], [500, 500]],
    'learning_rate': [0.001, 0.01]
}

model = KerasClassifier(build_fn = build_model, epochs=10)

# Create RandomizedSearchCV for padded data
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_distributions, n_iter=100, cv=3)
random_result = random_search.fit(X_train, Y_train)
```

## Determine Best Params

```python
# Get the results from the random search
results = random_search.cv_results_

# Create a DataFrame from the results
results_df = pd.DataFrame({
    'mean_test_score': results['mean_test_score'],
    'std_test_score': results['std_test_score'],
    'params': [str(p) for p in results['params']]
})

# You can also add individual columns for each hyperparameter by extracting them from the 'params' column
for key in param_distributions.keys():
    results_df[key] = results_df['params'].apply(lambda x: eval(x)[key])

# Sort by mean_test_score in descending order and take the top 10
results_df.sort_values(by='mean_test_score', ascending=False).head(10)
```

## Retrain with Defined Params So We Can See Loss Curves

```python
# Get the best parameters found by RandomizedSearchCV
best_params = random_search.best_params_

# Build a new model using these parameters (with padding)
model = build_model(**best_params)

# Fit the model to the data
history = model.fit(X_train, Y_train, epochs=10, validation_data=(X_test, Y_test))

# Plot the training and validation loss
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()
```

## Model Evaluation

Accuracy & Precision

```python
loss, accuracy, precision = model.evaluate(X_test, Y_test)

# Obtain predictions
y_pred = model.predict(X_test) # Corrected to use X_test instead of X_val

# Convert probabilities to class labels if necessary
y_pred = (y_pred > 0.5).astype(int)

# Calculate F1 score using Scikit-Learn's function
f1 = f1_score(Y_test, y_pred, average = 'macro') # Corrected to use Y_test instead of Y_val

# Print or otherwise use the F1 score
print("F1 Score:", f1)
print("Accuracy:", accuracy)
print("Precision:", precision)
```

Examine SHAP values to see what model learned

```python
# Ignore all FutureWarnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Ignore all tensorflow warnings 
tf.get_logger().setLevel('ERROR')

# Define the explainer
explainer = shap.GradientExplainer(model, X_train)

# Get the indices for one positive and one negative example
indices = [np.where(Y_test == 1)[0][0], np.where(Y_test == 0)[0][0]]

plt.rcParams['figure.dpi'] = 150

# Explain predictions of the model on the selected images
shap_values = explainer.shap_values(X_test[indices])

# Plot the feature attributions for each image
for i, idx in enumerate(indices):
    # Get the corresponding SHAP values for the index
    shap_value = shap_values[0][i]
    # Plot the SHAP values
    shap.image_plot(shap_value, np.squeeze(X_test[idx:idx+1]))
```
